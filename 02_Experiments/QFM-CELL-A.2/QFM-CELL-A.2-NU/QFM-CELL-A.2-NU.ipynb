```python
# ì •ëŸ‰ ì§€í‘œ ë¶„ì„ ì½”ë“œ ì‚½ì… :

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['font.family'] = 'AppleGothic'  # macOS ê¸°ë³¸ í•œê¸€ í°íŠ¸
matplotlib.rcParams['axes.unicode_minus'] = False  # ë§ˆì´ë„ˆìŠ¤ ê¸°í˜¸ ê¹¨ì§ ë°©ì§€

import numpy as np
from scipy.spatial.distance import cosine

# ì •ëŸ‰ ì§€í‘œ í•¨ìˆ˜ ì •ì˜
def shannon_entropy(counts):
    total = sum(counts.values())
    probs = [v / total for v in counts.values()]
    return -sum(p * np.log2(p) for p in probs if p > 0)

def distribution_vector(counts, keys):
    total = sum(counts.values())
    return np.array([counts.get(k, 0) / total for k in keys])

# ì‹œë®¬ë ˆì´í„° ì„¤ì •
backend = AerSimulator()
shots = 1024

# íšŒë¡œ A.2.1 - ê´€ì¸¡ ìœ ë„
qc_a21 = QuantumCircuit(2, 1)
qc_a21.h(0)
qc_a21.cx(0, 1)
qc_a21.measure(0, 0)

# íšŒë¡œ A.2.2 - ê°„ì„­ ì‚½ì…
qc_a22 = QuantumCircuit(2, 1)
qc_a22.h(0)
qc_a22.cx(0, 1)
qc_a22.x(1)
qc_a22.measure(0, 0)

# ì‹¤í–‰
job_a21 = backend.run(qc_a21, shots=shots)
job_a22 = backend.run(qc_a22, shots=shots)

# ê²°ê³¼ ìˆ˜ì§‘
counts_a21 = job_a21.result().get_counts()
counts_a22 = job_a22.result().get_counts()

# ì‹œê°í™” ë° ì €ì¥
fig1 = plot_histogram(counts_a21, title="QFM-CELL-A.2.1: ê´€ì¸¡ ìœ ë„")
fig1.savefig("result_a21.png")
plt.close(fig1)

fig2 = plot_histogram(counts_a22, title="QFM-CELL-A.2.2: ê°„ì„­ ì‚½ì…")
fig2.savefig("result_a22.png")
plt.close(fig2)

# ì •ëŸ‰ ì§€í‘œ ê³„ì‚° ë° ì¶œë ¥
entropy_a21 = shannon_entropy(counts_a21)
entropy_a22 = shannon_entropy(counts_a22)

all_keys = sorted(set(counts_a21) | set(counts_a22))
v1 = distribution_vector(counts_a21, all_keys)
v2 = distribution_vector(counts_a22, all_keys)
similarity = 1 - cosine(v1, v2)

print(f"ğŸ“Š Shannon Entropy:")
print(f"A.2.1 (ê´€ì¸¡ ìœ ë„): {entropy_a21:.4f}")
print(f"A.2.2 (ê°„ì„­ ì‚½ì…): {entropy_a22:.4f}")
print(f"\nğŸ“Š Cosine Similarity Between A.2.1 and A.2.2: {similarity:.4f}")

# ì—”íŠ¸ë¡œí”¼ ë¹„êµ ì‹œê°í™”
plt.bar(["A.2.1", "A.2.2"], [entropy_a21, entropy_a22])
plt.title("ìƒ¤ë…¼ ì—”íŠ¸ë¡œí”¼ ë¹„êµ (Shannon Entropy Comparison)")
plt.ylabel("Entropy")
plt.grid(True)
plt.savefig("entropy_comparison.png")
plt.show()
```
